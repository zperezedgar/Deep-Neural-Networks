```
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets

iris = datasets.load_iris()
```

```
idx  = iris.target<2        # primeras dos clases
y    = iris.target[idx]     # {0, 1}
#Y[1] = 1-Y[1]              # cambiamos a un dato su etiqueta
x    = iris.data[idx, :2]   # primeras dos características

plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], 
            x[:, 1], 
            c=y, 
            cmap=plt.cm.flag)
plt.xlabel('Longitud Sépalos')
plt.ylabel('Ancho Sépalos')
plt.show()
```

```
import keras
import numpy as np

import matplotlib.pyplot as plt
```

Una neurona con activación sigmoidal

```
from keras import models #modelos de NN
from keras import layers #modelos capas

logistica = models.Sequential()   #modelos NN tipo secuencial (no retroalimentación)

#agregamos una neurona a nuestra capa
#entran datos de dimensión dos mas un bias automático

logistica.add(layers.Dense(units= 1,                # numero de neuronas en la capa 
                    activation  = 'sigmoid',        # funcion de activacion = sigmoide
                    name        = 'neurona_unica',  # nombre de la capa  
                    input_shape = (2,)))            # forma de la entrada: (szIm, ) la otra 
                                                    # (dimension del dato, ¿tamano de lote?), 

logistica.summary()
```

```
#ahora damos los datos para entrenamiento (compilar)

logistica.compile(optimizer='adam', #buen método para entrenar
                  loss     ='binary_crossentropy', #función objetivo
                  metrics  =['accuracy']) #métrica para evaluar durante entrenamiento
```

```
#entrenamos
#si corremos este pedazo de código, inicia con los pesos obtenidos en el entrenamiento anterior, (si es el primero los inicia al azar??)

import time #para ver cuanto tarda
tic=time.time()

history = logistica.fit(x         = x, 
                        y         = y, 
                        epochs    = 1000, #revisamos los datos 1000 veces
                        shuffle   = True, #revolvemos para no repetir los conjuntos (batches)
                        batch_size= 20, #cada paso revisamos conjuntos de 20 en 20
                        verbose   = 0, #queremos ver progreso o no
                        validation_split=0.2, #de todos los datos usamos un subconjunto par avalidar (no para entrenar)
                        )

print('Tiempo de procesamiento (secs): ', time.time()-tic)
```

```
y_pred = logistica.predict(x).squeeze()

score = logistica.evaluate(x, y, verbose=2)
print('Test loss:',     score[0])
print('Test accuracy:', score[1])
```

```
#vemos cuanles datos clasificó correctamente
y_pred>0.5
```

```
plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], 
            x[:, 1], 
            c=((y_pred>0.5)+y)/2., 
            cmap=plt.cm.jet)
plt.xlabel('Longitud Sépalos')
plt.ylabel('Ancho Sépalos')
plt.show()
```
